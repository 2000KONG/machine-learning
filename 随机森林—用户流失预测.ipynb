### 第二步：拆分测试集与训练集

```python
Target=df['broadband']
data=df.iloc[:,1:-1] #取df里第二列至倒数第一列的所有数据
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, Target, 
                                    test_size=0.4, random_state=123)
```

### 第三步：决策树进行建模

```python
import sklearn.tree as tree
from sklearn.model_selection import GridSearchCV
param_grid={
    'criterion': ['entropy', 'gini'], # 树的深度评估指标
    'max_depth': [2, 3, 4, 5, 6, 7, 8], # 树的深度
    'min_samples_split': [4, 8, 12, 16, 20, 24, 28]
}
```

```python
clf=tree.DecisionTreeClassifier() # 定义一个分类树
clfcv=GridSearchCV(
    estimator=clf,
    param_grid=param_grid,
    scoring='roc_auc',  # 评价指标
    cv=4
)
```

```python
# 进行训练
clfcv.fit(x_train, y_train)
```

```python
print(clfcv.best_params_)
```

```python
# 调参，重新建模，训练，测试
clf2=tree.DecisionTreeClassifier(
    criterion='entropy', 
    max_depth=5,
    min_samples_split=12   
)
clf2.fit(x_train,y_train)
pred=clf2.predict(x_test)
pred
```

### 第四步：随机森林进行建模

```python
param_grid = {
    'criterion':['entropy','gini'],# 衡量标准
    'max_depth':[5, 6, 7, 8],    # 每棵决策树的深度
    'n_estimators':[11,13,15],  # 决策树个数 - 随机森林特有参数
    'max_features':[0.3,0.4,0.5], # 每棵决策树使用的变量占比 - 随机森林特有参数
    'min_samples_split':[4,8,12,16]  # 叶子的最小拆分样本量
}
```

```python
import sklearn.ensemble as ensemble

rfc=ensemble.RandomForestClassifier()
rfc_cv=GridSearchCV(
    estimator=rfc,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=4
)
rfc_cv.fit(x_train,y_train)
```

```python
rfc_cv.best_params_
```

```python
# 调参，重新训练和测试
rfc2=ensemble.RandomForestClassifier(
    criterion='entropy',
    max_depth= 8,
    max_features= 0.3,
    min_samples_split= 4,
    n_estimators=15
)
rfc2.fit(x_train,y_train)
```

```python
pred=rfc2.predict(x_test)
pred
```
