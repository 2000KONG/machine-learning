### 第二步：对oecd_bli ,gdp_per_capita 列表进行规整

```python
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli
#只要TOT的
```

```python
oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
```

```python
# 把country列作为索引
gdp_per_capita.set_index('Country',inplace=True)
```

## 第五步：模型选择

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
import sklearn.linear_model
```

```python
full_country_stats=pd.read_csv('country_states.csv')
full_country_stats
```

```python
#可以看到上图的散点图比较分散，可以删除几个比较偏离的值
remove_indices=[0, 1, 6, 8, 33, 34, 35]
keep_indices=list(set(range(36))-set(remove_indices))
sample_data = full_country_stats[['Country','GDP per capita' ,'Life satisfaction']].iloc[keep_indices]
missing_data = full_country_stats[['Country','GDP per capita', 'Life satisfaction']].iloc[remove_indices]
```

```python
sample_data
```

```python
missing_data
```

```python
# 两个dataframe的index都设置为Country
sample_data.set_index('Country',inplace=True)
missing_data.set_index('Country',inplace=True)
```

```python
sample_data
```

```python
full_country_stats.plot(kind='scatter',x='GDP per capita',
                        y='Life satisfaction',figsize=(5,3),c='b')
```

```python
x=sample_data['GDP per capita']
y=sample_data['Life satisfaction']
x_remove=missing_data['GDP per capita']
y_remove=missing_data['Life satisfaction']
plt.xlabel('GDP per capita')
plt.ylabel('Life satisfaction')
plt.plot(x,y,'bo',x_remove,y_remove,'ro')
```

#### 对sample_data 进行可视化

```python
sample_data.plot(kind='scatter',x='GDP per capita',
                        y='Life satisfaction',figsize=(5,3),c='b')
```

```python
# 以country为索引，对特殊值进行标记,(x,y)为图中写国家名的坐标
position_text={
    "Hungary": (5000, 1),
    "Korea": (18000, 1.7),
    "France": (29000, 2.4),
    "Australia": (40000, 3.0),
    "United States": (52000, 3.8),
}
```

```python
sample_data.plot(kind='scatter',x='GDP per capita',
                        y='Life satisfaction',figsize=(5,3),c='b')
plt.axis([0,60000,0,10])
#x，y坐标的取值范围
for country,pos_text in position_text.items():
    pos_data_x, pos_data_y = sample_data.loc[country]
    country = "U.S." if country == "United States" else country
    plt.annotate(country,xy=(pos_data_x, pos_data_y),xytext=pos_text,
               arrowprops=dict(facecolor='black', width=0.5, shrink=0.1, headwidth=5) )
    plt.plot(pos_data_x,pos_data_y,'ro')
plt.show()
```

通过这个图可以看出来，用直线模拟比较合理
